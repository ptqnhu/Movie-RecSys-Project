```python
mapping = {}
for index, row in genre_list.iterrows():
    mapping[row['id']] = row['name']

print(mapping)
```

```python
def replace_genre_ids(genre_id):
    """
    Replaces genre_ids with corresponding names based on mapping dictionary.

    Args:
        genre_id (str): The string representation of a genre_id

    Returns:
        str: The replaced genre name (e.g. Drama, Crime, etc.).
    """

    try:
        # Handle single value or lists efficiently
        if isinstance(genre_id, list):
            return ', '.join(mapping.get(value) for value in genre_id)
        else:
            return mapping.get(genre_id, genre_id)
    except KeyError:
        # Gracefully handle missing keys
        print(f'WARNING: Unknown genre ID: {genre_id}')
        return genre_id
```

```python
df['genre_ids'] = df['genre_ids'].apply(lambda x: eval(x))
```

```python
df['replaced_genre_ids'] = df['genre_ids'].apply(replace_genre_ids)
print(df)
```
